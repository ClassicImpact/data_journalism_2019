```python
import re
import sys

def refine(TEXT):
    EMOJI = re.compile('['
                        u"\U0001F600-\U0001F64F"  # emoticons
                        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                        u"\U0001F680-\U0001F6FF"  # transport & map symbols
                        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                        u"\U00010000-\U0010ffff"
                        ']+', flags=re.UNICODE)
    return EMOJI.sub(r'', TEXT)
```

```python
import nltk
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from nltk.sentiment.util import *
from nltk.sentiment.vader import SentimentIntensityAnalyzer


list_ex = ['I really like jennies voice bcoz it gives american vibes when she sings powerful,stable voice.','Jennies long hair my life although fake still 😂',
          'I was there, when they were performing see u later Jisoo and Jennie made a heart and the music started Jennie broke the heart and Jisoo’s face was sooo sad and cute, my heart melted']

sid_ex = SentimentIntensityAnalyzer()
sentimentcompound_ex = []
for i in list_ex:
    print(i)
    ss = sid_ex.polarity_scores(i)
    #for k in sorted(ss):
       # print('{0}: {1}, '.format(k, ss[k]), end='')
   # print()
    sorted(ss)
    sentimentcompound_ex.append(ss)
sentimentcompound_ex
```

```python
with open('random2000_eng4.txt', 'r',encoding='utf-8') as f:
    list_file = []
    for line in f:
        list_file.append(line)

sid = SentimentIntensityAnalyzer()
sentimentcompound = []
for i in list_file:
    print(i)
    ss = sid.polarity_scores(i)
    #for k in sorted(ss):
       # print('{0}: {1}, '.format(k, ss[k]), end='')
   # print()
    sorted(ss)
    sentimentcompound.append(ss)
```

```python
compounds = []
for i in sentimentcompound:
    a = i.get('compound')
    compounds.append(a)
    

pos = []
neg = []
neu = []
for i in compounds:
    if (i < -0.05):
        neg.append(i)
    elif (-0.05 <= i <= 0.05):
        neu.append(i)
    elif (i > 0.05):
        pos.append(i)


sum1 = 0
positive_words = {}
for i in pos:
    sum1 += i
positive_words['mean_value'] = sum1 / len(pos)
positive_words['ratio%'] = len(pos)*100 / len(compounds)
pos_words_ratio = positive_words.get('ratio%')
pos_words_value = positive_words.get('mean_value')

sum2 = 0
negative_words = {}
for i in neg:
    sum2 += i
negative_words['mean_value'] = sum2 / len(neg)
negative_words['ratio%'] = len(neg)*100 / len(compounds)
neg_words_ratio = negative_words.get('ratio%')
neg_words_value = negative_words.get('mean_value')

sum3 = 0
neutral_words = {}
for i in neu:
    sum3 += i
neutral_words['mean_value'] = sum3 / len(neu)
neutral_words['ratio%'] = len(neu)*100 / len(compounds)
neu_words_ratio = neutral_words.get('ratio%')
neu_words_value = neutral_words.get('mean_value')




```

```python
import string
import pandas as pd
import math
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

new_pos = int(np.round(pos_words_ratio/0.01))*0.01
new_neg = int(np.round(neg_words_ratio/0.01))*0.01
new_neu = int(np.round(neu_words_ratio/0.1))*0.1
x= ["Positive", "Negative", "Neutral"]
y = [new_pos, new_neg, new_neu]


for i, v in enumerate(y):
    plt.text(i, v+3, str(v), color='black',va = 'center', ha = 'center',fontweight='bold')
    plt.bar(x, y, width=0.5, facecolor = '#F2FF64', edgecolor='white')
plt.ylim(0,100)
plt.xlabel("Sentiment")
plt.ylabel("% of comments")
plt.title("Daily Vlog")
#plt.show()
plt.savefig("Sentiment_ratio%_category4.jpg")
```

```python
import numpy as np

unique_nouns = set()
for sent in tagged_sentences2:    
    for word, tag in sent:
        if tag in ['NNP', 'NNG']:
            unique_nouns.add(word)

unique_nouns = list(unique_nouns)
unique_nouns = [n for n in unique_nouns if n not in ['!!', '!!!', '영상', '언니']]

occurs = np.zeros([len(tagged_sentences2), len(unique_nouns)])
np.shape(occurs)
```

```python
noun_index = {noun: i for i, noun in enumerate(unique_nouns)}
noun_index
```

```python
for i, sent in enumerate(tagged_sentences2):
    for word, tag in sent:
        if tag in ['NNP', 'NNG']:
            try:
                index = noun_index[word]  # 명사가 있으면, 그 명사의 인덱스를 index에 저정
                occurs[i][index] = 1  # 문장 i의 index 자리에 1을 채워 넣는다.
            except:
                True

                
occurs[0]
```

```python
co_occurs = occurs.T.dot(occurs)
```

```python
co_occurs[0]
```

```python
import networkx as nx
graph = nx.Graph()

for i in range(len(unique_nouns)):
    for j in range(i + 1, len(unique_nouns)):
        if co_occurs[i][j] > 25:
            graph.add_edge(unique_nouns[i], unique_nouns[j])
```
